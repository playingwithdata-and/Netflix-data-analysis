## Data Visualization

```python
a,ax = plt.subplots(figsize = (10,5))
sns.countplot(data = df, x='type', palette = 'viridis', ax=ax)
plt.show()
```

```python
a,ax=plt.subplots(figsize=(10,5))
sns.countplot(data=df, x="rating",palette="plasma",ax=ax)
plt.xticks(rotation=90)
plt.show()
```

```python
df['rating'].value_counts().plot.pie(autopct='%1.1f%%',shadow=True,figsize=(10,8))
plt.show()
```

```python
date = df[["date_added"]].dropna()

date['year'] = date["date_added"].apply(lambda i: i.split(', ')[1])
date['month'] = date["date_added"].apply(lambda i: i.lstrip().split(' ')[0])
```

```python
plt.xticks(rotation=45)
sns.countplot(x=date["month"])
plt.show()
```
